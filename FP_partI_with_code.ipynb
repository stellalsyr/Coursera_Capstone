```python
### Create a function to get nearby venues within a pre-specified radius(I used 500m based on investor's requirement):
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
### See those venues found by Foursquare:
Irvine_venues = getNearbyVenues(names=df_Irvine['Neighborhood'],
                                   latitudes=df_Irvine['Latitude'],
                                   longitudes=df_Irvine['Longitude'], radius=500)
```

```python
### Need to convert the data to one-hot format for further use:
# one hot encoding
Irvine_onehot = pd.get_dummies(Irvine_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
Irvine_onehot['Neighborhood'] = Irvine_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [Irvine_onehot.columns[-1]] + list(Irvine_onehot.columns[:-1])
Irvine_onehot = Irvine_onehot[fixed_columns]

Irvine_onehot.head()
```

```python
### Visualize the one-hot data matrix grouped by neighborhood:
Irvine_grouped = Irvine_onehot.groupby('Neighborhood').mean().reset_index()
Irvine_grouped
```

```python
### View top 10 venues in each neighborhood by frequency:

num_top_venues = 10

for hood in Irvine_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = Irvine_grouped[Irvine_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
### To view the top 10 venues in each neighborhood more clearly by creating a dataframe for above information:
# Create a useful function:
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]

# Apply the function:
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = Irvine_grouped['Neighborhood']

for ind in np.arange(Irvine_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(Irvine_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted
```

Now the dataframe with the information of top 10 venues within each neighborhood in Irvine is ready.

#### Step 2: Use k-means method to cluster the neighborhoods into 5 clusters.

```python
# import library
from sklearn.cluster import KMeans

# set number of clusters
kclusters = 5

Irvine_grouped_clustering = Irvine_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(Irvine_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_
```

```python
Irvine_merged = df_Irvine

Irvine_merged['Cluster Labels'] = kmeans.labels_

Irvine_merged = Irvine_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

Irvine_merged 
```

#### Step 3: Examine those 5 clusters and find the best cluster which has Chinese Restuarant listed in the top 10 most common venues within the neighborhood (if listed in the top 3 most common venues would be even better), and the same time, hopefully not having coffee shops/cafe places listed in the top 10 most common venues in the same neighborhood.

Using the above criteria, the best cluster label is 0.
